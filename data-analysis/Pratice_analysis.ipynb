```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/Users/ayush/Documents/Data/data-analysis/Data Science Salary 2021 to 2023.csv'

# Read the CSV file
Salary_df = pd.read_csv(file_path,)
pd.set_option('display.max_Columns',None)
```

```python
Salary_df.isnull().sum()
```

```python
Salary_df.sample(5)
```

```python
# analysis of salary according to experience_level
avg_salary_by_exp = Salary_df.groupby('experience_level')['salary_in_usd'].mean()
avg_salary_by_exp.plot(kind='bar')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Experience Level')
plt.xticks(rotation=1)
plt.show()
```

```python
plt.figure(figsize=(16, 7))
Salary_df['Work_year'] = Salary_df['work_year'].astype(int)
# Create a line plot using Seaborn
sns.lineplot(data=Salary_df, x='Work_year', y='salary_in_usd')
plt.xlabel('Years')
plt.ylabel('Salary (USD)')
plt.title('Salary Changes Over Years (Line Plot)')

# Show the plot

plt.show()
```
