```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
penquins=sns.load_dataset('penguins')
penquins.head()
```

```python
penquins['bill_depth_mm'].plot()
```

```python
penquins['bill_depth_mm'].plot(linestyle='', 
                               marker='s', 
                               markersize=10,
                               alpha=.5, 
                               figsize=(10, 5),
                               markerfacecolor='yellow',
                               markeredgecolor='black',)
```

```python
penquins['bill_depth_mm'][:50].plot(kind='bar',figsize=(10,5),color='red')
```

```python
penquins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,5),color='purple',alpha=0.6)
```
